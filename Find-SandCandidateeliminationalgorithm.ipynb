3. Data pre-processing

a. Find the number of missing values.

```python
dt.loc[5,'Income'] = None
dt.loc[15,'Attendance'] = None
dt.loc[20, 'Experience'] = None
print("Number of missing values in each column:")
print(dt.isnull().sum())
```

b. replace missing values

```python
dt['Income'].fillna(dt['Income'].mean(),inplace=True)
dt['Attendance'].fillna(dt['Attendance'].median(),inplace=True)
dt['Experience'].fillna(dt['Experience'].mode(),inplace=True)
dt
```

c. Apply encoding techniques for categorical values:

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct= ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['Department'])], remainder='passthrough')
x=ct.fit_transform(dt)
ec = ct.named_transformers_['encoder'].get_feature_names_out(['Department'])
rc = dt.columns.difference(['Department'])
ac = np.concatenate([ec, rc])
x_en = pd.DataFrame(x, columns=ac)

print("After One-Hot Encoding for 'Department':")
print(x_en)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=dt.iloc[:,9]
y=le.fit_transform(y)
print(y)
```

Implementing the Find_S algorithm

```python
def findS(c, target):
    h=c[0].copy()
    for i, val in enumerate(c):
        if target[i]==1: #1 here is Yes
            for j in range(len(h)):
                if val[j]!=h[j]:
                    h[j]='?'
    return h
sh=findS(x, y)
print("Find-S Algorithm: ", sh)
```

```python
def candidateE(c,target):
    spec_h=[c[0]]
    gen_h=[['?' for _ in range(len(c[0]))]]
    for i, val in enumerate(c):
        if target[i]==1:
            for j in range(len(spec_h)):
                if spec_h[j] !=val[j]:
                    spec_h[j] ='?'
            gen_h=[g for g in gen_h if not all(g[i]=='?' or g[i] == val[i] for i in range(len(g)))]
        else:
            gen_h=[g for g in gen_h if not all(g[i] == val[i] or g[i] == '?' for i in range(len(g)))]
    return spec_h, gen_h
shc, ghc= candidateE(x,y)
print("Candidate Elimination Specific Hypothesis: ", shc)
print("Candidate Elimination Specific Hypothesis: ", ghc)
```

```python
concepts = np.array(dt.iloc[:, 0:-1])
target = np.array(dt.iloc[:, -1])

def candidateE(concepts, target):
    specific_h = concepts[0].copy()
    general_h = [["?" for _ in range(len(specific_h))] for _ in range(len(specific_h))]
    print("Initialization of specific_h and general_h")
    print("Specific Hypothesis:", specific_h)
    print("General Hypothesis:", general_h)
    for i, h in enumerate(concepts):
        if target[i] == "Yes":
            for x in range(len(specific_h)):
                if h[x] != specific_h[x]:
                    specific_h[x] = '?'
            general_h = [g for g in general_h if not all(g[k] == '?' or g[k] == h[k] for k in range(len(h)))]
        else: 
            for j in range(len(general_h)):
                for x in range(len(specific_h)):
                    if h[x] != specific_h[x]:
                        general_h[j][x] = specific_h[x] 
                    else:
                        general_h[j][x] = '?' 
            general_h = [g for g in general_h if g != ['?' for _ in range(len(specific_h))]]
        print(f"After processing {'positive' if target[i] == 'Yes' else 'negative'} instance: {h}")
        print("Specific Hypothesis:", specific_h)
        print("General Hypothesis:", specific_h)
    return specific_h, general_h
s_final, g_final = candidateE(concepts, target)
print("Final Specific_h:", s_final, sep="\n")
print("Final General_h:", s_final, sep="\n")
```
